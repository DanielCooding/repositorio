```python
import pandas as pd
url = 'https://raw.githubusercontent.com/nisston/disciplinaestruturadedados/main/data_temp.csv'
df1 = pd.read_csv(url)
lista1= df1['Temperature'].dropna().to_list()
def bubble_sort(lista):
  n =len(lista)
  for i in range(n):
    for j in range(0, n-i-1):
      if lista[j] > lista[j+1]:
        temp= lista[j]
        lista[j]= lista[j+1]
        lista[j+1]= temp
    return lista
lista_ordenada= bubble_sort(lista1)
top20=lista_ordenada[:20]
soma = 0
for i in top20:
  soma += i
media = soma/ 20
print(media)
print(lista_ordenada[:20])
```
